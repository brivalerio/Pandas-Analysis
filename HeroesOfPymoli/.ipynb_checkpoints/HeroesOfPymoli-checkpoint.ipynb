```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
purchases = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(purchases)
purchase_df.head(10) #Testing file read
```

## Player Count

* Display the total number of players


```python
players = len(purchase_df["SN"].unique())
total_players = pd.DataFrame({"Total Players": [players]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_df["Item ID"].unique())
average_price = purchase_df["Price"].mean()
number_of_purchases = len(purchase_df["Purchase ID"])
total_revenue = purchase_df["Price"].sum()

purchasing_analysis = pd.DataFrame({"Number of Unique Items":[unique_items], "Average Price":[average_price],
                                   "Number of Purchases":[number_of_purchases], "Total Revenue":[total_revenue]})

#Format clean up
purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].map("${:,.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map("${:,.2f}".format)

purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Dropping duplicate gender counts from same usernames
unique_genders = purchase_df.drop_duplicates(subset="SN", keep="first")

genders_total = unique_genders["Gender"].count()
male = unique_genders["Gender"].value_counts()["Male"]
female = unique_genders["Gender"].value_counts()["Female"]
other = unique_genders["Gender"].value_counts()["Other / Non-Disclosed"]

percentage_male = (male/genders_total) *100
percentage_female = (female/genders_total) *100
percentage_other = (other/genders_total) *100

demographics = pd.DataFrame({"": ["Male", "Female", "Other / Non-Disclosed"],
                            "Total Count": [male, female, other],
                             "Percentage of Players": [percentage_male, percentage_female, percentage_other]})

#Format clean up
demographics["Percentage of Players"] = demographics["Percentage of Players"].map("{:.2f}".format)

demographics = (demographics.set_index(''))

demographics.style.set_properties(**{'text-align': 'left'})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
groupby_gender = purchase_df.groupby(["Gender"])

purchase_count = groupby_gender["Purchase ID"].count()
purchase_value = groupby_gender["Price"].sum()
purchase_average = purchase_value/purchase_count

#Finding Avg Total Purchase per Person by unique SN
genders_total_x = unique_genders["Gender"].value_counts()
purchase_avg_players = purchase_value/genders_total_x

gender_purchases = pd.DataFrame({"Purchase Count": purchase_count,
                                "Average Purchase Price": purchase_average,
                                "Total Purchase Value": purchase_value,
                                 "Avg Total Purchase per Person": purchase_avg_players
                                })

#Format clean up
gender_purchases["Average Purchase Price"] = gender_purchases["Average Purchase Price"].map("${:,.2f}".format)
gender_purchases["Total Purchase Value"] = gender_purchases["Total Purchase Value"].map("${:,.2f}".format)
gender_purchases["Avg Total Purchase per Person"] = gender_purchases["Avg Total Purchase per Person"].map("${:,.2f}".format)

gender_purchases.style.set_properties(**{'text-align': 'left'})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,150]
age_group = ['<10', '10 - 14', '15 - 19', '20 - 24', '25 - 29', '30 - 34', '35 - 39', 'Over 40']

# purchasedf_copy1 = purchase_df.drop_duplicates(subset="SN", keep="first")
purchase_df[""] = pd.cut(purchase_df["Age"], bins, labels=age_group)

#purchase_df.drop_duplicates(subset="SN", keep="first")

binned_age = purchase_df.groupby([""])

age_count = binned_age["SN"].nunique()
total_count = purchase_df["SN"].nunique()
age_percentage = (age_count/total_count) *100
age_demographics = pd.DataFrame ({"Total Count": age_count,
                                 "Percentage of Players": age_percentage})

#Format clean up
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:.2f}".format)

age_demographics.style.set_properties(**{'text-align': 'left'})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
group_sn = purchase_df.groupby(["SN"])

purchase_count_sn = group_sn["Purchase ID"].count()
purchase_value_sn = group_sn["Price"].sum()
purchase_average_sn = group_sn["Price"].mean()

top_spenders = pd.DataFrame({"Purchase Count": purchase_count_sn,
                            "Average Purchase Price": purchase_average_sn,
                            "Total Purchase Value": purchase_value_sn})

#Sorting top spenders by Total Purchase Value in decending order
top_5_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)

#Format clean up
top_5_spenders["Average Purchase Price"] = top_5_spenders["Average Purchase Price"].map("${:,.2f}".format)
top_5_spenders["Total Purchase Value"] = top_5_spenders["Total Purchase Value"].map("${:,.2f}".format)

top_5_spenders.head().style.set_properties(**{'text-align': 'left'})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
groupby_items = purchase_df.set_index(["Item ID","Item Name"])#Setting new indices
groupby_items = purchase_df.groupby(["Item ID","Item Name"])

item_counts = groupby_items["Item ID"].count()
item_sum = groupby_items["Price"].sum()
item_price = item_sum / item_counts

pop_items = pd.DataFrame({"Purchase Count": item_counts,
                         "Item Price": item_price,
                         "Total Purchase Value": item_sum})

#Sorting by Purchase Count in decending order
most_pop_items = pop_items.sort_values("Purchase Count", ascending=False)

#Format clean up
most_pop_items["Item Price"] = most_pop_items["Item Price"].map("${:,.2f}".format)
most_pop_items["Total Purchase Value"] = most_pop_items["Total Purchase Value"].map("${:,.2f}".format)

most_pop_items.head().style.set_properties(**{'text-align': 'left'})
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable_items = pop_items.sort_values("Total Purchase Value", ascending=False)

#Format clean up
most_profitable_items["Item Price"] = most_profitable_items["Item Price"].map("${:,.2f}".format)
most_profitable_items["Total Purchase Value"] = most_profitable_items["Total Purchase Value"].map("${:,.2f}".format)

most_profitable_items.head().style.set_properties(**{'text-align': 'left'})
```
